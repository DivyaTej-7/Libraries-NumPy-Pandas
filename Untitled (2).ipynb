```python
import pandas as pd

df = pd.read_csv('C:\\Users\\DivyaTejAtluri\\Downloads\\retail_dataset .csv')

print("First 5 Rows:")
display(df.head())

print("\nDataset Info:")
df.info()
```

```python
print("Missing values per column:")
display(df.isnull().sum())

print("\nTotal missing values in dataset:")
print(df.isnull().sum().sum())
```

```python
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['Price'] = df['Price'].fillna(df['Price'].mean())

print("After filling missing values:")
display(df[['Quantity', 'Price']].isnull().sum())
```

```python
df = df.dropna(subset=['Product Category', 'Region'])

print("After dropping missing values:")
display(df.isnull().sum())
```

```python
df['Revenue'] = df['Quantity'] * df['Price']

print("Dataset with Revenue column:")
display(df.head())
```

```python
total_revenue = df['Revenue'].sum()

print("Total Revenue Generated:")
print(total_revenue)
```

```python
category_revenue = df.groupby('Product Category')['Revenue'].sum()

print("Revenue by Product Category:")
display(category_revenue)
```

```python
category_revenue = df.groupby('Product Category')['Revenue'].sum()

top_3 = category_revenue.sort_values(ascending=False).head(3)
bottom_3 = category_revenue.sort_values().head(3)

print("Top 3 Categories:")
display(top_3)

print("Bottom 3 Categories:")
display(bottom_3)
```

```python
region_revenue = df.groupby('Region')['Revenue'].sum()

print("Revenue by Region:")
display(region_revenue)

print("\nRegion with Highest Revenue:")
print(region_revenue.idxmax(), ":", region_revenue.max())

print("\nRegion with Lowest Revenue:")
print(region_revenue.idxmin(), ":", region_revenue.min())
```

```python
import numpy as np

# Fix date format issue
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Extract Month
df['Month'] = df['Date'].dt.month

# Monthly Revenue
monthly_revenue = df.groupby('Month')['Revenue'].sum()

print("Monthly Revenue:")
display(monthly_revenue)

# Statistics
mean_rev = np.mean(df['Revenue'])
median_rev = np.median(df['Revenue'])
std_rev = np.std(df['Revenue'])

print("\nRevenue Statistics:")
print("Mean:", mean_rev)
print("Median:", median_rev)
print("Standard Deviation:", std_rev)
```
